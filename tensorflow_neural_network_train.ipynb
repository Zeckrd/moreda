```python
import pandas as pd
import numpy as np
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

IMPORT DATASET

```python
df_movie = pd.read_csv('/trainDataset/dataset_movie.csv', nrows=600000).fillna(0)
df_user = pd.read_csv('/trainDataset/dataset_user.csv', nrows=600000).fillna(0)
df_rating = pd.read_csv('/trainDataset/dataset_rating.csv', nrows=600000).fillna(0)
```

df_movie movieid release year

```python
movie_feature = df_movie.shape[1]-2
user_feature = df_user.shape[1]-2
```

```python
user_feature
```

```python
movie_feature
```

scale dataset for training

```python
movie_train_unscaled = df_movie
user_train_unscaled = df_user
rating_train_unscaled = df_rating

scale_movie = StandardScaler()
scale_movie.fit(df_movie)
df_movie = scale_movie.transform(df_movie)

scale_user = StandardScaler()
scale_user.fit(df_user)
df_user = scale_user.transform(df_user)

scale_rating = MinMaxScaler()
scale_rating.fit(np.array(df_rating).reshape(-1,1))
df_rating = scale_rating.transform(np.array(df_rating).reshape(-1,1))
```

```python
print(np.allclose(movie_train_unscaled, scale_movie.inverse_transform(df_movie),equal_nan=True))
print(np.allclose(user_train_unscaled, scale_user.inverse_transform(df_user)))
```

```python
movie_train, movie_test= train_test_split(df_movie, test_size=0.2, random_state=42)
user_train, user_test= train_test_split(df_user, test_size=0.2, random_state=42)
rating_train, rating_test = train_test_split(df_rating, test_size=0.2, random_state=42)
```

model

```python
output = 32
movie_model = keras.models.Sequential([
    keras.layers.Dense(738, activation="relu"),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(output)
])

user_model = keras.models.Sequential([
    keras.layers.Dense(738, activation="relu"),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(output)
])
```

input for model

```python
input_user= tf.keras.layers.Input(shape=(user_feature,))
con_user = user_model(input_user)
con_user = tf.keras.layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1))(con_user)

input_movie = tf.keras.layers.Input(shape=(movie_feature,))
con_movie = movie_model(input_movie)
con_movie = tf.keras.layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1))(con_movie)

dot_product = tf.keras.layers.Dot(axes=1)([con_user, con_movie])

model = tf.keras.Model(inputs=[input_user, input_movie], outputs=[dot_product])
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.MeanSquaredError(),
    metrics=[tf.keras.metrics.MeanSquaredError()]
)
history = model.fit(
    [user_train[:,2:], movie_train[:,2:]], rating_train,
    batch_size = 256,
    epochs=40
)
```

```python
model.save('capstone.h5')
# model.save('capstone.keras')
```
